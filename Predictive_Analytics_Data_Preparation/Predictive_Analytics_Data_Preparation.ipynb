```
import pandas as pd
import io
import requests
url="https://raw.githubusercontent.com/stan587/Python-Portfolio/master/Data_Scraping_Project/rainforest_contract.csv"
source=requests.get(url).content
df=pd.read_csv(io.StringIO(source.decode('utf-8')))

df.columns.values[0] = "id"

# reversing the id since the first person signed up should be assigned with the smallest number
# in this case i'm starting the id at 1
df.id=1997-df.id
df
```

```
# Next step, I'm interested in deriving new fields where shows me the count of cap letters in first name and last name
# simple program for counting upper letters

df['fn_upper_count']=df['first_name'].str.count(r'[A-Z]')
df['ln_upper_count']=df['last_name'].str.count(r'[A-Z]')
df['fn_len']=df['first_name'].str.len()
df['ln_len']=df['last_name'].str.len()
df

```

```
from difflib import SequenceMatcher
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
df['fl_similarity']=''
df['label']=''

# Iteratively calculate fl_similarity, if error will continue and we'll manually examine what's the cause of the issue
for i,row in df.iterrows():
  try:
    df.fl_similarity[i]=similar(df.first_name[i], df.last_name[i])
  except Exception: 
    pass

# At this stage, we can see that we have what need to train the machine learning model,
# Next I'll export this and move on to labeling this data.


df.to_csv(r'./machine_learning_data_prep.csv',index=False)
df
```
